```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/rohit/Downloads/train.csv")
```

```python
print (df.shape)
```

```python
print (df.count())
```

```python
fig = plt.figure(figsize=(18, 6))
```

```python
plt.show()
```

```python
df.Survived.value_counts().plot(kind="bar", alpha=0.5)
```

```python
df.Survived.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
```

```python
plt.scatter(df.Survived, df.Age, alpha=0.1)
plt.title ("Age wrt Survived")
```

```python
df.Pclass.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plt.title("Class")
```

```python
for x in [1, 2, 3]:
    df.Age[df.Pclass == x].plot(kind="kde")
plt.title("Class wrt Age")
plt.legend(("1st", "2nd", "3rd"))
```

```python
df.Survived[df.Sex == "male"].value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plt.title("Men Survived")
```

```python
df.Survived[df.Sex == "female"].value_counts(normalize=True).plot(kind="bar", alpha=0.5, color="#FA0000")
plt.title("Female Survived")
```

```python
df.Sex[df.Survived == 1].value_counts(normalize=True).plot(kind="bar", alpha=0.5, color=["#FA0000", 'b'])
plt.title("Sex of Survived")
```

```python
for x in [1, 2, 3]:
    df.Survived[df.Pclass == x].plot(kind="kde")
plt.title("Survived wrt class")
plt.legend(("1st", "2nd", "3rd"))
```

```python
df.Survived[(df.Sex == "male") & (df.Pclass == 1)].value_counts(normalize="True").plot(kind="bar", alpha = 0.5)
plt.title("Rich men survived")
```

```python
df.Survived[(df.Sex == "male") & (df.Pclass == 3)].value_counts(normalize="True").plot(kind="bar", alpha = 0.5)
plt.title("Poor men survived")
```

```python
df.Survived[(df.Sex == "female") & (df.Pclass == 1)].value_counts(normalize="True").plot(kind="bar", alpha = 0.5, color="#FA0000")
plt.title("Rich women survived")
```

```python
df.Survived[(df.Sex == "female") & (df.Pclass == 3)].value_counts(normalize="True").plot(kind="bar", alpha = 0.5, color="#FA0000")
plt.title("Poor women survived")
```
